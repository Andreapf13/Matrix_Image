The calculate_element_destination_matrix function is going to receive the row position, element position and the original matrix.
First of all we are going to look for the indices of the neighbors. We are going to look for both the indexes of the original array
and the indexes.With this we will be able to
The i variable is going to be the one that stores the indices of the origin_matrix matrix.
The j variable is going to be the one that stores the indices of the lists that are in the origin matrix.
That's why we assign variables for the neighbors of origin_matrix and the neighbors of the given index of i.

```python
def calculate_element_destination_matrix(i,j,origin_matrix):
    value = origin_matrix[i][j]
    average = value
    count = 1 
    if i != 0:
        i_above = i - 1 
        j_above = j
        value_above = origin_matrix[i_above][j_above]
        average = average + value_above
        count = count + 1 
    if i != (len(origin_matrix) - 1 ):
        i_below =  i + 1 
        j_below = j
        value_below =  origin_matrix[i_below][j_below]
        average = average + value_below 
        count = count + 1 
    if j != 0:
        i_left = i
        j_left =  j - 1
        value_left =  origin_matrix[i_left][j_left]
        average = average + value_left
        count = count + 1 
    if j != len(origin_matrix[i]) - 1:
        i_right = i 
        j_right = j + 1 
        value_right = origin_matrix[i_right][j_right]
        average = average + value_right
        count = count + 1 
    return round((average / count),2)
```

The process_matrix function allows us to go from the origin matrix function with said average of its neighbors to the destination matrix.
To do this, we first pass an empty list and then go through the origin matrix with two nested for. The nested for allows us to pass
both as the positon row and the position element. In this way we will obtain the final value that will be the one that we add to our empty list.
What this function will allow us is to have the representation of the origin and destination matrix in a "prettier" way.
In other words, what we obtained previously was a print of a list of lists and what we obtain in this way is the closest thing
to the real representation of a matrix.


```python
def process_matrix(origin_matrix):
    destination_matrix = []
    for  i in range(len(origin_matrix)):
        i_destination = []
        for j in range(len(origin_matrix[i])):
            element_destination =  calculate_element_destination_matrix (i,j,origin_matrix)
            i_destination.append(element_destination)
        destination_matrix.append(i_destination)
    return destination_matrix


long_save = int(input("Tell me how long you want the original matrix to be "))
wide_save = int(input("Tell me how wide you want the original matrix to be "))

origin_matrix = numpy.random.randint(0,10,(long_save,wide_save)) #This allows us to create random numbers from 0 to 10.
print(origin_matrix)
```

```python
def matrix_representation(origin_matrix):
    a = ""
    for long in range(len(origin_matrix)):
        for wide in range(len(origin_matrix[-1])):
            a += str(origin_matrix[long][wide]) + "\t" 
        a += "\n"
    print("THIS IS ORIGIN MATRIX:" "\n" "\n" + a)
    process_matrix_representation = process_matrix(origin_matrix)
    b = ""
    for long in range(len(process_matrix_representation)):
        for wide in range(len(process_matrix_representation[-1])):
            b+= str(process_matrix_representation[long][wide]) + "\t" 
        b += "\n"   
    print("THIS IS DESTINATION MATRIX:" "\n" "\n"+ b)
matrix_representation(origin_matrix)
```
